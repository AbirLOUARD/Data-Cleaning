Import Packages

```python
import pandas as pd
```

Open data

```python
data = pd.read_csv("/content/sample_data/data_cleaning_challenge.csv")
data.shape
```

```python
data.head()
```

Drop the columns Unnamed 9 and 10

```python
drop = data.drop(columns=["Unnamed: 9", "Unnamed: 10"])
drop
```

Drop the values NaN 

```python
drop_nan = drop[drop["Row Type"].notna()]
drop_nan
```

```python
columns_values = []
counter = 0

for i in drop_nan["Row Type"]:
  if "first name" in i:
    counter += 1 
  columns_values.append(counter)

columns_values
```

Add an iteration column

```python
add_col = drop_nan
add_col["iteration"] = columns_values
add_col
```

```python
drop_col_row = add_col[add_col["Row Type"] != "Row Type"]
drop_col_row
```

```python
name = drop_col_row[drop_col_row["Row Type"].str.contains("first name")]
name
```

```python
name = name.loc[:,["Row Type", "Iter Number", "Power1", "iteration"]]
name.rename(columns={"Row Type": "First Name", "Iter Number":"Last Name", "Power1": "Date"}, inplace=True)
name["First Name"] = name["First Name"].str[12:]
name["Last Name"] = name["Last Name"].str[10:]
name["Date"] = name["Date"].str[5:]
name
```

```python
no_name = drop_col_row[~drop_col_row["Row Type"].str.contains("first name")]
no_name
```

Merge


```python
pd.merge(left=name, right=no_name, how="inner", on="iteration")
```
